#  Data Understanding: Patient Medical History Features

## Notebook Purpose

This notebook focuses on understanding the **patient medical history variables** in the dataset used for the **Triagegeist emergency triage prediction task**.

Medical history plays an important role in emergency medicine because **pre-existing conditions significantly influence patient risk, disease severity, and triage priority**. Understanding these variables helps in building models that better approximate **clinical decision making during emergency triage**.

---

## Features Covered in This Notebook

This analysis explores the **14 medical history (`hx_`) columns**, which indicate whether a patient has previously been diagnosed with specific medical conditions.

The columns analyzed include:

- `hx_hypertension`
- `hx_diabetes_type1`
- `hx_diabetes_type2`
- `hx_asthma`
- `hx_copd`
- `hx_heart_failure`
- `hx_atrial_fibrillation`
- `hx_ckd`
- `hx_liver_disease`
- `hx_malignancy`
- `hx_obesity`
- `hx_depression`
- `hx_anxiety`
- `hx_dementia`

Each variable is a **binary indicator**:

- **0** → No known history of the condition  
- **1** → Patient has a prior diagnosis of the condition

These variables provide important information about **baseline patient vulnerability and comorbidity burden**, which can influence emergency severity outcomes.

---

## Goal of This Notebook

The goal of this notebook is to explore the **distribution of patient medical history variables** by examining the **value counts** of each `hx_` column.

Each `hx_` feature is a **binary indicator** representing whether a patient has a history of a specific medical condition.

By analyzing the value counts, we aim to:

- Understand the **prevalence of each medical condition** in the dataset.
- Verify that the features are **properly encoded as binary variables (0 = No history, 1 = History present)**.
- Gain an initial understanding of the **baseline health conditions of the patient population**.

This step is part of the **initial data understanding phase** before moving to deeper exploratory analysis and feature engineering.

### 3️ What is Type 1 Diabetes

Type 1 diabetes is a **chronic autoimmune disease** where:

- The immune system destroys **insulin-producing beta cells** in the pancreas.
- The body produces **little or no insulin**.
- Patients must take **insulin injections for life**.

It usually develops **earlier in life** compared to Type 2 diabetes.

---

### 4️ Why This Feature Is Important in Emergency Triage

Patients with Type 1 diabetes are at risk of **acute metabolic emergencies** that can rapidly become life-threatening.

Common emergency risks include:

#### Diabetic Ketoacidosis (DKA)
A serious condition where:
- Blood sugar becomes extremely high
- The body produces **ketones**
- Blood becomes acidic

Symptoms may include:
- Rapid breathing
- Vomiting
- Confusion
- Severe dehydration

DKA is a **true medical emergency**.

---

### 5️ Clinical Importance in Triage

Emergency clinicians consider **Type 1 diabetes a high-risk condition** because:

- Blood sugar can become unstable quickly
- Infection or stress can trigger **DKA**
- Patients may deteriorate **rapidly**

Example:

A patient presenting with:

- abdominal pain
- rapid breathing

This can **increase triage urgency**.

### 3️ What is COPD

- **Chronic bronchitis** (long-term airway inflammation)
- **Emphysema** (damage to lung air sacs)

Common symptoms include:

- Shortness of breath  
- Chronic cough  
- Excess mucus production  
- Wheezing  
- Reduced oxygen intake

COPD is most commonly caused by:

- long-term smoking  
- air pollution exposure

- Acute COPD exacerbation  
- Respiratory failure  
- Severe pneumonia  
- Hypoxia (low oxygen levels)

- Fluid buildup in the lungs

### 3️ What is Chronic Kidney Disease (CKD)

**Chronic Kidney Disease** is a long-term condition where the **kidneys gradually lose their ability to filter waste and excess fluids from the blood**.

The kidneys normally:

- Remove toxins from the blood  
- Maintain electrolyte balance  
- Regulate blood pressure  
- Control fluid levels in the body  

When CKD develops:

- Waste products accumulate in the body  
- Fluid balance becomes unstable  
- Electrolyte abnormalities may occur  

Common causes include:

- diabetes  
- hypertension  
- kidney infections  
- autoimmune diseases  

---

### 4️ Why This Feature Is Important in Emergency Triage

Patients with CKD are **medically fragile** because their kidneys cannot properly maintain internal body balance.

They are at higher risk for serious conditions such as:

- **electrolyte imbalances (e.g., high potassium)**
- **fluid overload**
- **acute kidney failure**
- **severe infections**
- **cardiovascular complications**

Electrolyte abnormalities, especially **high potassium (hyperkalemia)**, can cause **dangerous heart rhythm disturbances**.

---
